### Monte Carlo - Predicting Gross Profit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20 
interations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, interations)
rev
```

```python
plt.figure(figsize = (15,6))
plt.plot(rev)
plt.show()
```

```python
## the above figure lies in the band of = 190 = 170 + 20, 170, 150 = 170 - 20
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize = (15,6))
plt.plot(COGS)
plt.show
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize = (15,6))
plt.plot(Gross_Profit)
plt.show
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```

```python
Gross_Profit.mean()
```

```python
Gross_Profit.std()
```
